5. Write a program to do: A dataset collected in a cosmetics shop showing
details of customers and whether or not they responded to a special offer
to buy a new lip-stick is shown in table below. (Use library commands)
According to the decision tree you have made from the previous training
data set, what is the decision for the test data: [Age < 21, Income = Low,
Gender = Female, Marital Status = Married]?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('Datasets/Lipstick.csv')

df.drop(['Id'],axis=1,inplace=True)
print(df)
input = {'Age' : '<21', 'Income' : 'Low',
'Gender' : 'Female', 'Ms' : 'Married'}
input = pd.DataFrame(input, index=[0])
input
```

```python
df_encoded = pd.get_dummies(df, columns=['Age', 'Income', 'Gender', 'Ms'])
df_encoded

```

```python
input_encoded = pd.get_dummies(input,columns=['Age', 'Income', 'Gender', 'Ms'])
input_encoded
```

```python

model_input = df_encoded.copy(deep=True)
model_input.drop(model_input.index, inplace=True)
model_input.drop(['Buys'],axis=1, inplace=True)
model_input.loc[len(model_input)] = 0
model_input.update(input_encoded,overwrite=True)
model_input
```

```python
model = DecisionTreeClassifier()
model.fit(df_encoded.drop('Buys', axis=1),df_encoded['Buys'])
```

```python
model.predict(model_input)
```

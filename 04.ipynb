4. Write a program to do: A dataset collected in a cosmetics shop showing
details of customers and whether or not they responded to a special offer
to buy a new lip-stick is shown in table below. (Implement step by step
using commands - Dont use library) Use this dataset to build a decision
tree, with Buys as the target variable, to help in buying lipsticks in the
future. Find the root node of the decision tree.


```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('Datasets\Lipstick.csv')
df.drop('Id', axis=1, inplace=True)
df
```

```python
total_instances = len(df)
class_counts = df['Buys'].value_counts()
print(class_counts)
total_entropy = -sum((count/total_instances) * math.log2(count/total_instances) for count in class_counts)
print(total_entropy)
```

```python
for col in df.drop('Buys',axis=1).columns:
    groups = df.groupby(col)
    print()
    group_entropy = 0

    for group_name, group_df in groups:
        group_instances = len(group_df)
        group_count = group_df['Buys'].value_counts()
        group_entropy += (group_instances/total_instances) * (-sum((count/group_instances) * math.log2(count/group_instances) for count in group_count)) 

    print(f"Entropy for {col} is {group_entropy}")

    info_gain = total_entropy - group_entropy
    print(f"Information Gain for {col} is {info_gain}")

```

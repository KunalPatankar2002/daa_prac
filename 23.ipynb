attribute age. From the frequency table you have obtained, calculate the
information gain of the frequency table while splitting on Age. (Use step
by step Python/Pandas commands)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
dict = {
    "Age": ["Young", "Young", "Middle", "Old", "Old", "Old", "Middle", "Young", "Young", "Old", "Young", "Middle", "Middle", "Old"],
    "Income": ["High", "High", "High", "Medium", "Low", "Low", "Low", "Medium", "Low", "Medium", "Medium", "Medium", "High", "Medium"],
    "Married": ["No", "No", "No", "No", "Yes", "Yes", "Yes", "No", "Yes", "Yes", "Yes", "No", "Yes", "No"],
    "Health": ["Fair", "Good", "Fair", "Fair", "Fair", "Good", "Good", "Fair", "Fair", "Fair", "Good", "Good", "Fair", "Good"],
    "Class": ["No", "No", "Yes", "Yes", "Yes", "No", "Yes", "No", "Yes", "Yes", "Yes", "Yes", "Yes", "No"]
}
df = pd.DataFrame(dict)
df
```

```python
total_instances = len(df)
class_counts = df['Class'].value_counts()
print(class_counts)
total_entropy = -sum((count/total_instances) * math.log2(count/total_instances) for count in class_counts)
print(total_entropy)

# Step 2: Compute the entropy after splitting based on 'Age'
age_df = df.groupby('Age')
age_entropy = 0
for group_name, group_df in age_df:
    group_instances = len(group_df)
    group_count = group_df['Class'].value_counts()
    age_entropy += (group_instances/total_instances) * (-sum((count/group_instances) * math.log2(count/group_instances) for count in group_count)) 

print(age_entropy)
info_gain = total_entropy - age_entropy
print(info_gain)

# # Display the results
print("Entropy of the entire dataset:", total_entropy)
print("Entropy after splitting on 'Age':", age_entropy)
print("Information Gain:", info_gain)
```

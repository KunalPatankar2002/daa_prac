```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='pclass', data=titanic, palette='viridis')
plt.title('Passenger Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='pclass', hue='survived', data=titanic, palette='muted')
plt.title('Survival Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='pclass', y='age', data=titanic, palette='pastel')
plt.title('Age Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='sex', data=titanic, palette='deep')
plt.title('Passenger Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='sex', hue='survived', data=titanic, palette='Set2')
plt.title('Survival Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='pclass', y='age', hue='survived', data=titanic)
plt.title('Survival Distribution by Age and Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.legend(title='Survived', loc='upper left', labels=['No', 'Yes'])
plt.show()
```
